## Part A – Application area review

As modern cities - especially capital cities - are expanding continuously, controlling traffic to ensure a steady flow is becoming more and more of a concern for several reasons. It affects individual people commuting to work or to other responsibilities by taking up more of their time. By keeping running vehicles on the road for longer, it hurts the environment through air and noise pollution, as well as leading to the use of more fuel through idling. We will be looking at a number of different ways artificial intelligence can be applied to this problem domain with the aim of reducing congestion.

A journal by Reddy et al. (2023) suggests using Internet of Things devices, such as cameras and sensors, to control traffic signals rather than change them using timers. The proposed system uses computer vision to identify and count the number of vehicles coming from each direction to determine how the traffic signals should change. They were able to achieve real-time data processing, making this a viable solution. This is similar to how humans are often used as traffic controllers; however, their system could run continuously and without any human interaction, making it more efficient.

Wu et al. (2013) propose a traffic sign detection system. It uses a convolutional neural network that is trained on a dataset of traffic sign images. Each image in the training set belongs to one of three classes. During training, the model detects patterns in the images and extracts features from them. Once trained, their system was able to classify new images of traffic signs accurately.

Manikonda, Yerrapragada and Annasamudram (2011) introduce a so-called “Intelligent Traffic Management System” that uses radio-frequency identification tags found in vehicles, and devices that read these tags along roads. These readings are able to determine the average speed at which the vehicles are moving. The system then uses Dijkstra’s algorithm on the collected data to determine the quickest route between a given point and a destination. Drivers are meant to access this system through an interface within their vehicles, allowing them to pick the best route and thus avoid congestion.

Khan, Adnan and Iqbal (2022) discuss using artificial intelligence to quickly identify accidents and alert the authorities, who will be able to aid the situation. This will help avoid any potential secondary accidents and also reduce further congestion. Their solution combines object detection with region-based convolutional neural networks to detect and then report these incidents.

Mi and Yang (2022) have identified that road slippage caused by road icing makes up a significant portion of highway accidents, which can lead to traffic issues. Their solution proposes the use of various types of sensors to detect the presence of ice on the road and the temperature of the area, as well as cameras that capture the road surface to prevent incorrect icing detections. The collected data is then classified and used in artificial intelligence pattern recognition systems to determine if icing has indeed occurred, in which case it alerts the appropriate services.

Nechoska, Nechkoska and Duma (2022) present a solution that uses computer vision to automatically collect traffic data in certain locations and analyse traffic patterns with machine learning techniques, making it more accurate and cost-efficient than manual human observation and prediction. The data can then be used to create a “traffic simulation model”, which can give valuable information to city planning experts.

Alshriem (2020) mentions how neural networks have often been used to detect traffic violations. Similarly to the above studies, the traffic data is collected using various sensors next to roads, analysed by a separate system, and then using neural networks it is determined if such an incident has occurred. This method eliminates the delay that comes with human reports.

Fedchenkov et al. (2018) propose a system that uses artificial intelligence to suggest the best parking location based on where a vehicle is, making it quicker to find parking space and, in turn, reducing the number of vehicles on the road trying to look for parking space manually. It uses Internet of Things devices placed around parking lots to determine how many empty spaces there are. The system would also know the location of the vehicle, allowing it to calculate the distance from all nearby parking spaces. It would also have the pricing for each parking lot, further helping the drivers.

Yasdi (1999) explores a system that uses historical time-series traffic data with recurrent neural networks to forecast future congestion. The system uses loop detectors to collect traffic data, identifying the volume of vehicles passing over. The paper has also evaluated its model using the mean squared error loss function method, which resulted in a value of 0.003, indicating that it performs well.

## Part B – Compare and evaluate AI techniques

### Traffic Prediction with Neural Networks

Neural networks can be used to predict future traffic conditions by analysing traffic data. They are able to process large amounts of historical traffic data to detect patterns and make predictions about future traffic flow. The data includes things like timestamps, traffic volume, traffic speed, etc. They can also able to make predictions in real-time. The predictions can then be used within traffic management systems for various purposes (Passow et al., 2013).

#### _Strengths_

Neural networks are suitable for identifying complex patterns from large datasets, as shown by Ma, Song and Li (2020), which would otherwise be a difficult task to do manually. They can also detect nonlinear relationships from the datasets (Bonmati, 2023).

As mentioned by Povey, Zhang and Khudanpur (2014), neural networks are capable of parallel training, allowing them to process large amounts of data. This can speed up the process of training a model.

Marullo et al. (2021) talk about neural networks’ ability to adapt by introducing new training data and updating their capabilities. This is important because when it comes to traffic prediction, the existing historical data may get outdated if traffic patterns change over time.

#### _Weaknesses_

Neural networks are at risk of overfitting, as discussed by Lawrence and Giles (2000), meaning it would work well with the training dataset, but would fail to accurately analyse new data. This can happen since traffic conditions may involve a lot of parameters. Techniques like dropout or early stopping can be used to prevent overfitting (Bonmati, 2023).

Depending on the amount of training data and the complexity of the system, training neural networks can require significant computational power and can result in lengthy training time, as shown by Livni, Shalev-Shwartz and Shamir (2014).

If the learning rate of a neural network model is too fast or too slow, it can cause problems (Bonmati, 2023); therefore, it is important to find the appropriate learning rate. With the former, it could converge too quickly and give sub-optimal results, whereas with the latter, convergence would take too long.

#### _Input & output, data availability, setup time, time to produce results_

The input data for such systems would be traffic data, such as the number of vehicles passing by a location at specific times and the speed of said vehicles, as shown by Tedjopurnomo et al. (2020). The expected output would be a prediction of future traffic conditions at different times around the same locations.

Traffic prediction using neural networks would, more often than not, use historical data that was gathered beforehand. Although neural networks are capable of processing data in real-time as well, provided that there’s a separate system that also collects said data in real-time, as demonstrated by Huang, Zhu and Siew (2006).

As data collection is not necessarily involved in a traffic prediction system, we will not consider it when talking about setup time. As mentioned in the weaknesses section, the model training time is the biggest factor when dealing with neural networks. The training time - as well as the validation and testing - will depend on the hardware capabilities of the system where the model is trained. As mentioned in in the strengths section, neural networks can perform parallel training, which speeds up the training phase. Dogaru and Dogaru (2017) go over methods using GPU acceleration techniques to further speed up the training process.

Once the model has been trained, the time to produce results would depend on the prediction model's complexity as well as the input data.

### Route Optimisation with Search Algorithms

Route Optimisation using artificial intelligence search algorithms and Internet of Things devices is a method to find the most optimal route between certain locations. It involves the use of a variety of devices, for example, cameras or location trackers that act as sensors to collect information about individual vehicles and/or traffic conditions around them, such as the whereabouts of a car, its speed, the number of vehicles waiting at a junction, road conditions etc. The collection of such data can happen in real-time or over time, depending on the system. As Khanna et al. (2018) say, the accumulated information usually ends up being very large. The different data gathered can be used to create heuristics, which aid the search algorithms in finding the optimal route (Basukoski, 2023).

#### _Strengths_

According to Khanna et al. (2018), such systems are better at finding the best route since they consider an assortment of variables, such as path length, travel time, congestion history, etc., instead of only relying on any one of them. Considering so many factors manually, without the help of artificial intelligence, would be significantly more time-consuming for humans.

Another advantage brought up by Hu et al. (2020) is that such systems are able to accomplish real-time route planning. Through real-time data collection, the system can adjust its parameters and re-evaluate the most optimal path from the current location. This is important because, in the case of a lengthy route, traffic conditions can change, which could make the originally planned route less optimal.

Hu et al. (2020) also mention that by finding the most optimal transit route, we can minimise the overall cost of the journey by reducing things such as fuel and time consumption. Based on this, we can also assume that over time, road maintenance costs will also be reduced due to vehicles spending less time on roads.

#### _Weaknesses_

Rohunen, Heikkilä and Heikkilä (2014) talk about the privacy concerns that come with the mass collection of traffic data. Individual people may be concerned about having their location data collected and stored, especially since they would not know if any third parties would get access to said data, using it for a different purpose. Some of these systems rely on this data for their functionality therefore, it is unavoidable that it gets collected, which makes it disadvantageous from a user perspective.

Another potential disadvantage is the fact that many artificial intelligence search algorithms rely on heuristics to be optimal, as shown by Korf (1999). This means that with a lack of heuristics, or even improper heuristics, the algorithms can lead to sub-optimal solutions.

Korf (1999) also explains that the space complexity of some search algorithms, i.e. how much data they store in memory, might be high if the search tree is large. This means that the route optimisation system might have to run powerful and expensive hardware.

#### _Input & output, data availability, setup time, time to produce results_

As detailed in the above studies, the input data would be information gathered by the IoT sensors, such as location, speed, video (of traffic conditions), temperature, etc. The expected output would be the optimal route - a series of locations connected to each other - to take from a starting point to a destination.

Data availability depends on whether a system uses historical data or data collected in real-time; the former system would have to collect data over time, which is a process in itself, whereas with the latter, it would be available right away.

The setup time will first be affected by data preprocessing. As mentioned by Barapatre and Vijayalakshmi (2017), this is usually a time-consuming process. However, it is needed to ensure that the algorithms are able to work with the data efficiently and produce the correct output. Search algorithms do not need to be trained like machine learning models; therefore, we can assume that their setup time would not be significant.

As mentioned in the strengths section, these systems are able to produce results in real time, making them a viable choice.

### Traffic Image Classification

Image classification is an artificial intelligence technique that is used to classify new images using a deep convolutional neural network. Such a model is trained on a large dataset of images with corresponding labels. It then makes predictions about unseen images, by determining what class they belong to, based on pre-defined classes in the training data (Rawat and Wang, 2017). Deep convolutional neural networks can also learn to extract features from the training data by detecting patterns, making them efficient for image classification tasks (Jogin et al., 2018).

#### _Strengths_

Convolutional neural networks are capable of transfer learning. Ribani and Marengoni (2019) explain how not having a smaller dataset can be a problem when trying to train a model from scratch. Transfer learning allows the use of a pre-trained model to be re-trained on a custom dataset, which reduces the training time required.

Adding pooling layers to convolutional neural networks, with techniques such as max pooling and average pooling, can be used to downsample the input data (Psarrou, 2023). By reducing the spatial dimension of the input, they reduce the model’s training time and help with computational costs (Zafar et al., 2022).

Liu (2018) explains the efficient feature extraction capabilities of convolutional neural networks, aiding their accuracy. They are able to learn the hierarchy of features by first identifying edges, dark spots, etc., and ending up detecting the whole image.

Aghdam, Heravi and Puig (2016) mentioned that certain models such as ‘ConvNet’ have also managed to surpass human performance when it comes to image classification tasks.

#### _Weaknesses_

Li et al. (2016) explain how convolutional neural networks can be very resource-intensive and can require powerful hardware to be trained on. This also leads to high energy consumption.

Convolutional neural networks require large amounts of data to train, as Krichen (2023) explains. This means that extensive data collection needs to be done before implementing such a system.

#### _Input & output, data availability, setup time, time to produce results_

The input data for training an image classification model would be a large dataset of images alongside corresponding class labels. Once trained, the input data for inference would be images similar to the training data but without any labels. The output data would be the predicted classes of the new/unseen data (Psarrou, 2023).

As mentioned before, training an image classification model requires a large amount of labelled data; therefore, the unavailability of data could result in sub-optimal training.

Performing pre-processing procedures may take time during the setup process; however, it is necessary to ensure that the model learns efficiently, as mentioned by Pal and Sudeep (2016). Training convolutional neural networks is a resource-heavy process, and they need to be trained on large datasets; therefore, they may lead to lengthy training, which would affect the setup time.

The time to produce results will largely depend on how complicated the convolutional neural network model is, the hardware specifications of the system and the resolution of the images.

### Description of the input

We will be using the GTSRB - German Traffic Sign Recognition Benchmark dataset from Kaggle: https://www.kaggle.com/datasets/meowmeowmeowmeowmeow/gtsrb-german-traffic-sign.

The dataset contains 51839 images of 43 types of traffic signs. It is split into 39209 training images and 12630 testing images.

The Meta directory contains one example image for each class, with a corresponding Meta.csv file, but we will be plotting example images directly from the training set, therefore /Meta/ and Meta.csv were deleted.

We will use an additional file from a preprocessed version of the original dataset named Traffic Signs Preprocessed: https://www.kaggle.com/datasets/valentynsichkar/traffic-signs-preprocessed?select=label_names.csv.

We are only using the 'label_names.csv' file, which contains the actual sign name of each class. The implementation could still work without this file, however each class would only have a numerical value when visualising the data. The label names are used purely for the sake of clarity for these visualisations.

The data has been downloaded manually, and extracted into the directory of this notebook. The 'Train' and 'Test' directories have been gitignored because the large number of images causes issues.

#### Some pre-processing needs to be done:
* The data is only split into a training and testing set, therefore we will be splitting the training set into a training and validation set
* There are different images sizes in the dataset, therefore we will be resizing all images to 32x32 for better performance (Saponara and Elhanashi, 2021)
* We will convert images and labels into numpy arrays for efficient computation (Van Der Walt, Colbert and Varoquaux, 2011)
* Finally, we will convert the class labels to categorical format, since they work better with neural network classification tasks (Hancock and Khoshgoftaar, 2020)

### Prototype

#### Importing libraries

```python
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

#### Disabling GPU
This was needed on my system because of a CUDA error preventing tensorflow from training the model.

It disables GPU accelerated training, forcing the use of the CPU (making it a slower process).

It needs to be done before importing tensorflow

```python
os.environ["CUDA_VISIBLE_DEVICES"] = ""
```

#### Importing tensorflow
If viewing the notebook in Visual Studio Code with the Pylance extension, the import warnings about tensorflow.keras packages is an issue with Pylance.

If all packages are installed, the imports work correctly and the warnings can be ignored.

```python
from tensorflow.keras.utils import to_categorical # type: ignore
from tensorflow.keras.models import Sequential # type: ignore
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout # type: ignore
```

#### Read the class labels file

```python
csv_file_path = './label_names.csv'
label_names = pd.read_csv(csv_file_path)
print(label_names)
```

#### Create a dictionary to map ClassId to SignName

```python
class_dict = dict(label_names[['ClassId', 'SignName']].values)
print(class_dict)
```

#### Parse training data into arrays of images and labels
Also counting the number of images for each class and converting data to numpy arrays

```python
train_images = []
train_labels = []

# Dictionary to keep track of the number of images in each class
class_counter = { class_id: 0 for class_id in class_dict.keys() }

train_data = pd.read_csv('./Train.csv')

for _, row in train_data.iterrows(): # first iterator is named _ because it's not used
    with Image.open('./' + row['Path']) as image:
        image = image.resize((32, 32))
        image = np.array(image)
        class_number = int(row['ClassId'])
        train_images.append(image)
        train_labels.append(class_number)
        class_counter[class_number] += 1 # Update the counter for the current class

train_images = np.array(train_images)
train_labels = np.array(train_labels)

print(train_images.shape)
print(train_labels.shape)
```

#### Logging the number of images for each class in the training set

```python
print("Number of images for each class in the training set:")
for class_id, count in class_counter.items():
    class_name = class_dict[class_id]
    print(f"{class_name}: {count} images")
```

#### Visualising the number of images for each class in the training set

We can see that the number of images in the training set differ a lot across classes.

This means that the dataset is imbalanced, and could lead to some incorrect predictions. However, we will see that despite this, the model will still perform well overall.

```python
class_names = list(class_dict.values())
class_counts = list(class_counter.values())

plt.figure(figsize=(12, 8))
plt.barh(class_names, class_counts, color='skyblue')
plt.xlabel('Number of Images')
plt.title('Number of Images for Each Class in the Training Set')
plt.gca().invert_yaxis()
plt.show()
```

#### Splitting the data into training and validation sets

```python
train_images, val_images, train_labels, val_labels = train_test_split(train_images, train_labels, test_size=0.2, random_state=0)
print(train_images.shape, train_labels.shape)
print(val_images.shape, val_labels.shape)
```

#### Parse test data into arrays of images and labels

```python
test_data = pd.read_csv('./Test.csv')

test_labels = test_data["ClassId"].values
test_image_paths = test_data["Path"].values

test_images = []

for path in test_image_paths:
    with Image.open('./' + path) as image:
        image = image.resize((32, 32))
        image = np.array(image)
        test_images.append(image)

test_images = np.array(test_images)
test_labels = np.array(test_labels)

print(test_images.shape)
print(test_labels.shape)
```

#### Logging the number of images in each set

```python
print('Number of images in training set: ' + str(len(train_images)))
print('Number of images in validation set: ' + str(len(val_images)))
print('Number of images in testing set: ' + str(len(test_images)))
```

#### Plotting one random image from each class in the training set

```python
unique_classes = np.unique(train_labels)

num_rows = 7
num_cols = 7
plt.figure(figsize=(15, 15))

for i, class_id in enumerate(unique_classes):
    class_indices = np.where(train_labels == class_id)[0] # Get indices for the current class
    random_index = random.choice(class_indices)

    plt.subplot(num_rows, num_cols, i + 1)
    plt.imshow(train_images[random_index])
    plt.title(class_dict[class_id], fontsize=10)
    plt.axis('off')

plt.show()
```

#### Conversion to categorical format

```python
train_labels_cate = to_categorical(train_labels, num_classes=len(class_dict))
val_labels_cate = to_categorical(val_labels, num_classes=len(class_dict))

print(train_labels_cate)
print(val_labels_cate)
```

#### Building the model

We build the model using a structure proposed by Mehta, Paunwala and Vaidya (2019), since it was designed for a similar dataset.

```python
model = Sequential([
    Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(filters=128, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(units=2000, activation='relu'),
    Dropout(0.3),
    Dense(units=500, activation='relu'),
    Dropout(0.3),
    Dense(units=len(class_dict), activation='softmax')  # Output layer with number of classes
])
```

#### Compiling the model and showing the summary
We use the Adam optimiser, as Mehta, Paunwala and Vaidya (2019) demonstrated it performing well on traffic sign classification.

We use categorical cross entropy loss, because of the categorical format of the data.

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

#### Training the model

```python
history = model.fit(train_images, train_labels_cate, epochs=10, validation_data=(val_images, val_labels_cate))
```

#### Plotting accuracy and loss change over epochs

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

#### Plotting 30 random images from the test set, with their correct labels

```python
plt.figure(figsize=(15, 15))

# Get random indices for the testing set
random_indices_testing = np.random.choice(len(test_images), size=30, replace=False)

for i, random_index in enumerate(random_indices_testing):
    plt.subplot(5, 6, i + 1)
    plt.imshow(test_images[random_index])
    plt.title(f"Class: {class_dict[test_labels[random_index]]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

#### Making the predictions for the test set
Calculates the percentage of prediction accuracy.

Calculates precision, recall and F1-score for each class.

```python
prediction = np.argmax(model.predict(test_images), axis=1)
print('Test Data accuracy: ', accuracy_score(test_labels, prediction) * 100)

report = classification_report(test_labels, prediction, target_names=[class_dict[i] for i in range(len(class_dict))])
print("Classification Report:\n", report)
```

#### Visualises a confusion matrix of the predictions as a heatmap

```python
cm = confusion_matrix(test_labels, prediction)
plt.figure(figsize=(15, 15))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_dict.values(), yticklabels=class_dict.values())
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

#### Shows a bar chart of the precision, recall and F1-score for each class

We are not plotting the support, as the other values are in percentage form and the support is the number of instances in each class, meaning it would distort the bar chart.

```python
# Re-calculates the classification report, but now as a dictionary to be used for the chart
report = classification_report(test_labels, prediction, 
                               target_names=[class_dict[i] for i in range(len(class_dict))], output_dict=True)

plt.figure(figsize=(20, 3))
class_indices = range(len(class_dict))
bar_width = 0.2

plt.bar(class_indices,
        [report[class_name]['precision'] for class_name in class_dict.values()],
        width=bar_width, label='Precision')

plt.bar([index + bar_width for index in class_indices],
        [report[class_name]['recall'] for class_name in class_dict.values()],
        width=bar_width, label='Recall')

plt.title('Precision, Recall and F1-Score for each class')
plt.xlabel('Class')
plt.ylabel('Score')
plt.xticks(class_indices, class_dict.values(), rotation=45, ha='right')
plt.legend()
plt.show()
```
